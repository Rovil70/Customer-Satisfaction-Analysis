```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"D:\Unified Mentor Intership\Data Science projects\Customer Satisfaction\Data\customer_support_cleaned.csv")
```

Distribution of Customer Satisfaction Rating

```python
sns.countplot(x='Customer Satisfaction Rating', data=df)
plt.title('Customer Satisfaction Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

Satisfaction by Ticket Type

```python
sns.boxplot(x='Ticket Type', y='Customer Satisfaction Rating', data=df)
plt.xticks(rotation=45)
plt.title('Satisfaction by Ticket Type')
plt.show()
```

Satisfaction by Ticket Priority

```python
sns.boxplot(x='Ticket Priority', y='Customer Satisfaction Rating', data=df)
plt.title('Satisfaction by Ticket Priority')
plt.show()
```

Channel-wise Satisfaction

```python
sns.violinplot(x='Ticket Channel', y='Customer Satisfaction Rating', data=df)
plt.title('Satisfaction by Channel')
plt.show()
```

Correlation heatmap for numerical features

```python
numeric_cols = ['Customer Age', 'Response Delay (hrs)', 'Resolution Delay (hrs)', 'Customer Satisfaction Rating']
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
